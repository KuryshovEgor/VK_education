```python
class Node:
    def __init__ (self, data):
        self.next = None
        self.prev = None
        self.data = data
        
        
class spisok:
    def __init__ (self):
        self.head = None
        self.tail = None
    def append (self, data):       #добавить в конец
        new_node = Node(data)
        if self.head == None: 
            self.head = new_node
            self.tail = new_node
            return
        else:
            self.tail.next = new_node
            new_node.prev = self.tail
            self.tail = new_node
            return
    def display(self):
        current = self.head
        if not current:
            print("Список пуст.")
            return
        sp = np.array ([])
        while current:
#            print(current.data)
            sp = np.append(sp, current.data)
            current = current.next
        return sp
```

```python
#проверить цикличность
def hasCircle (head):
    if head == None or head.next == None:
        return False
    slow = head
    fast = head.next
    i=0
    while slow != fast:
        if fast == None or fast.next == None:
            return False
        slow = slow.next
        fast = fast.next.next
    return True
```

```python
#односвязный список
class Node:
    def __init__ (self, data):
        self.next = None
        self.data = data
        
        
class spisok:
    def __init__ (self):
        self.head = None
        self.tail = None
    def append (self, data):       #добавить в конец
        new_node = Node(data)
        if self.head == None: 
            self.head = new_node
            self.tail = new_node
            return
        else:
            self.tail.next = new_node
            self.tail = new_node
            return
    def display(self):
        current = self.head
        if not current:
            print("Список пуст.")
            return
        sp = np.array ([])
        while current:
#            print(current.data)
            sp = np.append(sp, current.data)
            current = current.next
        return sp
            

```

```python
#развернуть список
def reverseLinkedList(spis):
    prev = None
    current = spis.head
    while (current != None):
        next = current.next
        current.next = prev
        prev = current
        current = next
    spis.head = prev
    return spis


def test_reverseLinkedList():
    t = spisok()
    t.append(1)
    t.append(2)
    t.append(3)
    t.append(4)
    assert all(reverseLinkedList(t).display()) == all([4, 3, 2, 1])
```

```python
#слияние двух списков
def merge (A, B):
    pointer1 = A.head
    pointer2 = B.head
    while pointer1.next !=None and pointer2 != None:
        if pointer1.data > pointer2.data:      #случай когда первый элемент первого больше чем первый (и не только) элемент второго
            B.head = pointer2.next
            pointer2.next = pointer1
            A.head = pointer2
            pointer1 = pointer2
            pointer2 = B.head
        elif pointer2.data < pointer1.next.data:
            B.head = pointer2.next
            pointer2.next = pointer1.next
            pointer1.next = pointer2
            pointer1 = pointer2
            pointer2 = B.head
        else:
            pointer1 = pointer1.next
    if pointer2 != None:
        pointer1.next = pointer2
    return A

def test_merge():
    t = spisok()
    t.append(6)
    t.append(8)
    t.append(10)
    t.append(11)

    p = spisok()
    p.append(1)
    p.append(2)
    p.append(3)
    p.append(7)
    assert all(merge(t, p).display()) == all([1, 2, 3, 6, 7, 8, 10, 11])
```

```python
#Середина списка
def middleNode(head):
    slow = fast = head
    while fast != None and fast.next != None:
        slow = slow.next
        fast = fast.next.next
    return slow



def test_middleNode():
    t = spisok()
    t.append(1)
    t.append(2)
    t.append(3)
    t.append(4)
    t.append(5)
    assert middleNode(t.head).data == 3
```

```python
#Уалить узел по значению
def removeElements(A, val):
    dummy = Node(0)
    dummy.next = A.head
    prev = dummy
    cur = A.head
    while (cur != None):
        if (cur.data == val): 
            prev.next = cur.next
            if cur == A.head:
                A.head = cur.next
        else:
            prev = cur
        cur = cur.next

    return A


def test_removeElements():
    t = spisok()
    t.append(6)
    t.append(6)
    t.append(8)
    t.append(10)
    t.append(6)
    t.append(11)
    assert all(removeElements(t, 6).display()) == all([8, 10, 11])

```

```python
#Является ли одна строка частью другой
from queue import *
def isSubsequence(a, b):
    q = Queue()
    for el in a:
        q.put(el)
       
    for el in b:
        if q.queue[0] == el:
            q.get()
            if q.empty():
                break
    return q.empty()


def test_isSubsequence():
    a = "abs"
    b = "abbss"
    assert isSubsequence(a, b) == True
```

```python
#Полиндром
def isPalindrome(s):
    left = 0
    right = len(s) - 1

    while (left < right):
        if (s[left] != s[right]):
            return False
        left+=1
        right-=1

    return True


def test_isPalindrome():
    s = "abssba"
    assert isPalindrome(s) == True
```
