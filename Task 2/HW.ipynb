```python
import numpy as np
import ipytest
ipytest.autoconfig()
#односвязный список
class Node:
    def __init__ (self, data):
        self.next = None
        self.data = data
        
        
class spisok:
    def __init__ (self):
        self.head = None
        self.tail = None
    def append (self, data):       #добавить в конец
        new_node = Node(data)
        if self.head == None: 
            self.head = new_node
            self.tail = new_node
            return
        else:
            self.tail.next = new_node
            self.tail = new_node
            return
    def display(self):
        current = self.head
        if not current:
            print("Список пуст.")
            return
        sp = np.array ([])
        while current:
            sp = np.append(sp, current.data)
            current = current.next
        return sp
    
#слияние двух списков
def merge (A, B):
    pointer1 = A.head
    pointer2 = B.head
    while pointer1.next !=None and pointer2 != None:
        if pointer1.data > pointer2.data:      #случай когда первый элемент первого больше чем первый (и не только) элемент второго
            B.head = pointer2.next
            pointer2.next = pointer1
            A.head = pointer2
            pointer1 = pointer2
            pointer2 = B.head
        elif pointer2.data < pointer1.next.data:
            B.head = pointer2.next
            pointer2.next = pointer1.next
            pointer1.next = pointer2
            pointer1 = pointer2
            pointer2 = B.head
        else:
            pointer1 = pointer1.next
    if pointer2 != None:
        pointer1.next = pointer2
    return A

def test_merge():
    t = spisok()
    t.append(6)
    t.append(8)
    t.append(10)
    t.append(11)

    p = spisok()
    p.append(1)
    p.append(2)
    p.append(3)
    p.append(7)
    assert all(merge(t, p).display()) == all([1, 2, 3, 6, 7, 8, 10, 11])
    
ipytest.run()
```
